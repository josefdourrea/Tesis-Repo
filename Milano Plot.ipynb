```python
import geopandas as gpd
import matplotlib.pyplot as plt
import os
#import earthpy as et

# plot data inline
plt.ion() 
```

```python
%matplotlib inline
plt.rcParams['figure.figsize'] = (30, 30)
```

```python
milano = gpd.read_file('/media/jose/DATA/Dataset Telco/Examples/milano-grid.geojson')
```

```python
milano.head()
```

```python
LIM010102 = '/media/jose/DATA/Dataset Telco/DBT2012_STRATO01_E0/LIM_010102.shp'
LIM010103 = '/media/jose/DATA/Dataset Telco/DBT2012_STRATO01_E0/LIM_010103.shp'
LIM010104 = '/media/jose/DATA/Dataset Telco/DBT2012_STRATO01_E0/LIM_010104.shp'
LIM010105 = '/media/jose/DATA/Dataset Telco/DBT2012_STRATO01_E0/LIM_010105.shp'

map_LIM010102 = gpd.read_file(LIM010102)
map_LIM010103 = gpd.read_file(LIM010103)
map_LIM010104 = gpd.read_file(LIM010104)
map_LIM010105 = gpd.read_file(LIM010105)
```

```python
map_LIM010102.head(10)
```

```python
fig, ax = plt.subplots(figsize = (20,20))

map_LIM010102.plot(color='green',
                   edgecolor = 'black',
                   ax = ax,
                   legend = 'Peatonal',
                   alpha=.5)

map_LIM010103.plot(color='yellow',
                   edgecolor = 'black',
                   ax = ax,
                   legend = 'Cicloruta',
                   alpha=.5)

map_LIM010104.plot(color='blue',
                   edgecolor = 'black',
                   ax = ax,
                   legend = 'Via Principal',
                   alpha=.5)

map_LIM010105.plot(color='red',
                   edgecolor = 'black',
                   ax = ax,
                   legend = 'Via Secundaria',
                   alpha=.5)

ax.set_title('Movilidad y Transporte\nMilano')
#ax.set_axis_off()
plt.axis('equal');
```

```python
map_LIM010102_wgs84 = map_LIM010102.to_crs({'init': 'epsg:4326'})
map_LIM010103_wgs84 = map_LIM010103.to_crs({'init': 'epsg:4326'})
map_LIM010104_wgs84 = map_LIM010104.to_crs({'init': 'epsg:4326'})
map_LIM010105_wgs84 = map_LIM010105.to_crs({'init': 'epsg:4326'})
```

```python

fig, ax = plt.subplots(figsize = (20,20))
ax.set_title('Movilidad y Transporte\nMilano')
ax.set_axis_off()
plt.axis('equal');

map_LIM010102_wgs84.plot(color='green',
                         edgecolor = 'black',
                         ax = ax,
                         legend = 'Peatonal',
                         alpha=.2)

map_LIM010103_wgs84.plot(color='yellow',
                         edgecolor = 'black',
                         ax = ax,
                         legend = 'Cicloruta',
                         alpha=.2)

map_LIM010104_wgs84.plot(color='blue',
                         edgecolor = 'black',
                         ax = ax,
                         legend = 'Via Principal',
                         alpha=.2)

map_LIM010105_wgs84.plot(color='red',
                         edgecolor = 'black',
                         ax = ax,
                         legend = 'Via Secundaria',
                         alpha=.2)

milano.plot(color='black',
            edgecolor = 'black',
            ax = ax,
            alpha=.1)

ax.set_title('Movilidad y Transporte\nMilano')
ax.set_axis_off()
plt.axis('equal');
```
